```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize Graph
plt.figure(figsize = (5,3), dpi = 100)

#Line 1

# Keyword Argument Notation
# plt.plot(x,y, label = '2x', color = 'red', linewidth = 1, linestyle = '--',
# marker = '.', markersize = '8', markeredgecolor = 'black')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'b.--',label = '2x')

### LINE NUMBER 2

# select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r-', label = 'x^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')


# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph', fontdict={'fontname' : 'Comic Sans Ms', 'fontsize' : 18})

# X and Y labels
plt.xlabel('x axis')
plt.ylabel('y axis')


# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi = 300)

# Show plots
plt.show()
```

Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

plt.figure(figsize = (5,3), dpi = 100)


bars = plt.bar(labels, values, color=['red', 'green', 'blue'])

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.title('BAR CHART', fontdict={'fontname' : 'Comic Sans Ms', 'fontsize' : 18} )

plt.savefig('barchart.png', dpi=300)

plt.show()
```
